Dream house

Write a program that calculates how long it would take you to save enough money to pay the down payment for your dream house. You will ask the user for their annual salary, the percent of the salary they’d want to save, and the cost of their dream home.

You should assume that the down payment is 25% of the total cost of the home. The time should be expressed in months. Hint: annual salary / 12 will give you the monthly salary. 

```python
curr_ann_sal = round(float(input("What's your annual salary?  ")),2)
percent_sal = round(float(input('What percent of your salary will you save? ')),2)
cost = round(float(input('How much does your dream home cost? ')),2)
sem_ann_raise = round(float(input("What's the semi-annual raise percentage? ")),2)
inv_ann_ret = round(float(input("What's the investment annual return? ")),2)
down_pay = round(cost* (25/100),2)

saving = 0
month = 1
while saving < down_pay:
    mon_sal = curr_ann_sal/12
    saving = saving+(mon_sal)*(percent_sal/100)  
    if saving >= down_pay:
        break
    month += 1

years = int(month/12)
mon = int(month%12)

print("You would have to work for %d months to afford the downpayment for your dream house. %d months or %d years and %d months"%(month,month,years,mon))
```
